```python
import pandas as pd
df = pd.read_csv('reports/predictions_brocacef.csv')
df.head()
```

```python
df.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace infinite values with NaN, then drop them
df['remaining_jaren_p_70'].replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(subset=['remaining_jaren_p_70'], inplace=True)

# Calculate the histogram
counts, bin_edges = np.histogram(df['remaining_jaren_p_70'], bins=10)

# Print bin categories and value counts
print("Bin Edges:", bin_edges)
print("Counts:", counts)

# Plot the histogram
plt.hist(x=df['remaining_jaren_p_70'], bins=10)
plt.xlabel('Remaining Years at 70th Percentile')
plt.ylabel('Frequency')
plt.title('Histogram of Remaining Years at 70th Percentile')
plt.show()
```

```python
from lifelines import KaplanMeierFitter

kmf = KaplanMeierFitter()

# Fit the Kaplan-Meier model
kmf.fit(durations=df['tenure'], event_observed=df['Status_dienstverband_Uit dienst'])

# Plot the survival function
kmf.plot_survival_function()
plt.title('Kaplan-Meier Survival Curve')
plt.xlabel('Time (years)')
plt.ylabel('Survival Probability')
plt.show()
```

```python
# Check for outliers in tenure
plt.boxplot(df['tenure'])
plt.title('Boxplot of Tenure')
plt.ylabel('Years')
plt.show()

# Check the distribution of tenure
plt.hist(df['tenure'], bins=20)
plt.title('Histogram of Tenure')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.show()
```
